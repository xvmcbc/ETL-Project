```python
import pandas as pd
import requests
import json
from sqlalchemy import create_engine

Path = "C:/Users/karsa_000/Desktop/etl-project/winemag-data-130k-v3.csv"
wine_csv_df = pd.read_csv(Path, encoding="UTF'8")
wine_csv_df.head()
```

```python
json_file = "C:/Users/karsa_000/Desktop/etl-project/winemag-data-130k-v3.json"
wine_json_df = pd.read_json(json_file)
wine_json_df.head()
```

```python
complete_wine_df = pd.merge(wine_json_df, wine_csv_df, on = "title")

complete_wine_df.head()
```

```python
complete_wine_df= complete_wine_df.drop(columns="region_2")
complete_wine_df
```

```python
complete_wine_df= complete_wine_df.dropna()
```

```python
complete_wine_df.drop_duplicates(subset = "title", inplace = True)
complete_wine_df.head()
```

```python
complete_wine_df.insert(0, 'id', range(1, 1 + len(complete_wine_df)))

complete_wine_df.head()
```

```python
rds_connection_string = "postgres:191919@localhost:5432/wine_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
complete_wine_df.to_sql(name = "complete_wine", con = engine, if_exists = "append",index = False)
```
